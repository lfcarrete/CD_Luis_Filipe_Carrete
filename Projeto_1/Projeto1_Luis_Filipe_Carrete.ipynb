## Pergunta (s) a ser respondida

Existe um aumento no investimento estrangeiro de países que passaram a fazer parte da União Europeia?

## Explicação da hipótese e mecanismo

Com a adesão de um país a União Europeia ele passa a ter o Euro, a moeda centralizada da União Europeia e como resultado acaba tendo um aumento em investimentos estrangeiros por ter uma taxa mais baixa de conversão de câmbio, ou seja é mais barato para investidores que utilizam o euro investir em países que também utilizam o euro. Também reduz a diferença de preço entre países deixando o preço de produtos e serviços muito próximos por toda a Europa o que é conhecido como transparência de preços.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
from numpy import arange
from numpy import percentile
import numpy as np

dados = pd.read_excel('Dados_invext.xlsx')
```

```python
ue_dados = dados.loc[(dados.UE==1)]

ue_subiu = (ue_dados.Subiu.value_counts(sort=False, normalize=True)*100).round(decimals=1)


merco_dados = dados.loc[(dados.Merco==1)]

merco_subiu = (merco_dados.Subiu.value_counts(sort=False, normalize=True)*100).round(decimals=1)

fig = plt.figure(figsize=(20, 6))

plt.subplot(131)
plot = ue_subiu.plot(kind='bar',color=('b'))
plt.title("Países que subiram com a entrada na UE")
plt.ylabel("Frequência relativa")
plt.xlabel("0 = Não Subiu & 1 = Subiu")

plt.subplot(132)
plot = merco_subiu.plot(kind='bar',color=('green'))
plt.title("Países que subiram com a entrada no Mercosul")
plt.ylabel("Frequência relativa")
plt.xlabel("0 = Não Subiu & 1 = Subiu")
plt.show()
```

```python
ue_euro = ue_dados.loc[(ue_dados.Euro==1)]

ue_euro_subiu = (ue_euro.Subiu.value_counts(sort=False, normalize=True)*100).round(decimals=1)

ue_n_euro = ue_dados.loc[(ue_dados.Euro==0)]

ue_n_euro_subiu = (ue_n_euro.Subiu.value_counts(sort=False, normalize=True)*100).round(decimals=1)

fig = plt.figure(figsize=(20, 6))

plt.subplot(131)
plot = ue_euro_subiu.plot(kind='bar',color=('b'))
plt.title("Países que subiram que utilizam Euro")
plt.ylabel("Frequência relativa")
plt.xlabel("0 = Não Subiu & 1 = Subiu")


plt.subplot(132)
plot = ue_n_euro_subiu.plot(kind='bar',color=('r'))
plt.title("Países que subiram que não utilizam Euro")
plt.ylabel("Frequência relativa")
plt.xlabel("0 = Não Subiu & 1 = Subiu")

plt.show()
```

```python
faixa = np.arange(-7,15,1)

dados_e = dados.loc[(dados.UE==1)&(dados.Euro==1)]
dados_n = dados.loc[(dados.UE==1)&(dados.Euro==0)]

fig = plt.figure(figsize=(25, 5))
plt.subplot(131)
plot = dados_e[2010].plot.hist(bins=faixa,title='Dados 2010 Moeda Euro', density=True, color='b', alpha=0.5)
plt.ylabel('Densidade')
plt.xlabel('Investimento Externo/PIB')

plt.subplot(132)
plot = dados_n[2010].plot.hist(bins=faixa,title='Dados 2010 Outras Moedas', density=True, color='r', alpha=0.5)
plt.ylabel('Densidade')
plt.xlabel('Investimento Externo/PIB')

plt.show()
```
